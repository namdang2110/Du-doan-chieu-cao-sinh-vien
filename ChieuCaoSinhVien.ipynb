```python
import pandas as pd
import numpy as np

# k55 = pd.read_csv("D:\\Khoa hoc du lieu\\data_k55kmt.csv", index_col='Date', parse_dates=True)
k55 = pd.read_csv("D:\\Khoa hoc du lieu\\data_k55kmt.csv")
```

```python
k55.head()
```

```python
from sklearn.linear_model import LinearRegression
# Drop any rows with null values
k55.dropna(axis=0, how='any', inplace=True)

x = k55['Can Nang'].values.reshape(-1, 1)
y = k55['Chieu Cao']

model = LinearRegression(fit_intercept=False)
model.fit(x, y)
k55['Du doan'] = model.predict(x)
```

```python
k55[['Chieu Cao', 'Du doan']].plot(alpha=0.5);
```

Theo bảng dự đoán trên, đường "Du doan" luôn đi gần sát với đường "Chieu cao". Vậy nếu có giá trị dự đoán nào vượt xa ra khỏi đường "Chieu cao" đã được nhập sẵn, chứng tỏ giá trị chiều cao nhập vào có sự gian lận.
